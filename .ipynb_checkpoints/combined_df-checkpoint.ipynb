```python
# -----------------------------
# Import Necessary Libraries
# -----------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import odeint
# Import helper
from plot_utils import set_plot_style, set_spines_black
set_plot_style()
```

## Model 1

```python
# Parameters (Original)
s = 13000                 # cells/day
d = 0.0412               # /day
p = 0.1245               # /day
a = 0.18                 # /day
b = 2e-9                 # /cells
g = 2.019e7              # cells
m = 3.422e-10            # day^{-1} * cells^{-1}
n = 1.101e-7             # day^{-1} * cells^{-1}


# Parameters (Scaled)
alpha = p/d               
beta  = (m*g)/d            
gamma = a/d               
delta = (a*b*g)/d     
eps = (n*s)/d**2      
```

```python
## Define the ODEs and Generate the training data
def model1(state, t):
    x, y = state
    dx_dt = 1 - x + alpha*x*y/(1 + y) - beta*x*y
    dy_dt = gamma*y - delta*y**2 - eps*x*y
    return [dx_dt, dy_dt]

# Initial condition
state0 = [0.1, 0.1]
# Time grid
t0 = 0.0
dt = 0.01
tf = 50.0
t = np.arange(t0, tf + dt, dt)

# Integrate
x_train1 = odeint(model1, state0, t)
print(f'\nThe shape of the Training Data is: {x_train1.shape}')
```

```python
df_1 = pd.DataFrame(data=x_train1, columns=["E", "T"], index=t)
df_1.to_csv("model_1.csv")
```

## Model 2

```python
# Parameters (Original)
r2 = 0.18         # 1/day
mu2 = 0.03        # 1/day
mu3 = 10          # 1/day
p1 = 0.1245       # 1/day
p2 = 5            # 1/day
a = 1             # 1/day
b = 1e-9          # 1/cell
g1 = 2e7          # cells
g2 = 1e4          # cells
# g3 = 1e3          # cells
s1 = 10000        # cells/day
s2 = 38000        # cells/day

# Parameters (Scaled)
alpha = mu2/r2
beta = p1/r2
gamma = b*g2
delta = a*s1/(g2*r2**2)
eps = s2/(r2*g1)
mu = mu3/r2
nu = p2*s1/(g1*r2**2)
```

```python
## Define the ODEs and Generate the training data
def model2(state, t):
    x, y, z = state
    dx_dt = 1 - alpha*x + beta*x*z/(1 + z)
    dy_dt = y - gamma*y**2 - delta*x*y/(1 + y)
    dz_dt = eps - mu*z + nu*x*y/(1 + y)
    return [dx_dt, dy_dt, dz_dt]
# Initial condition
state0 = [0.1, 0.1, 0.1]
# time grid
t0 = 0.0
dt = 0.01
tf = 50.0
t = np.arange(t0, tf + dt, dt)

# Integrate
x_train2 = odeint(model2, state0, t)
print(f'\nThe shape of the Training Data is: {x_train2.shape}')
```

```python
df_2 = pd.DataFrame(data=x_train2, columns=["E", "T", "I"], index=t)
df_2.to_csv("model_2.csv")
```

## Model 3

```python
# Parameters 
a = 0.18                      # day−1
b = 1 / 30000                 # (cells/μl)^−1
gamma_c = 0.005              # unused
gamma_e = 0.0001             # (cells/μl)^−1 day−1
g = 100                      # cells/μl
kn = 0.001                   # day−1
alpha_n = 100
se = 36                      # (cells/μl)/day
de = 0.06                    # day−1
sn = 43.2                    # (cells/μl)/day
dn = 0.04                    # day−1

# Scaled Parameters
alpha = se/sn
beta = (alpha_n*kn)/de
gamma = (gamma_e*g)/de
delta = a/de
eps = (a*b*g)/de
mu = (gamma_c*sn)/(de)**2
nu1 = dn/de
nu2 = kn/de
```

```python
## Define the ODEs and Generate the training data
def model3(state, t):
    x, y, z = state
    dx_dt = alpha - x + beta*z*y/(1 + y) - gamma*x*y
    dy_dt = delta*y - eps*y**2 - mu*x*y
    dz_dt = 1 - nu1*z - nu2*z*y/(1 + y)
    return [dx_dt, dy_dt, dz_dt]

# Initial condition
state0 =  [0.1, 0.5, 0.1] 
# time grid
t0 = 0.0
dt = 0.01
tf = 50.0
t = np.arange(t0, tf + dt, dt)

# Integrate
x_train3 = odeint(model3, state0, t)
print(f'\nThe shape of the Training Data is: {x_train3.shape}')
```

```python
df_3 = pd.DataFrame(data=x_train3, columns=["E", "T", "N"], index=t)
df_3.to_csv("model_3.csv")
```

```python
# --------------------------------------------------------------------------------
#   Training data from all the three models has been generated.
# --------------------------------------------------------------------------------
```

```python
# --------------------------------------------------------
#  Load all the CSV files
# --------------------------------------------------------
df1 = pd.read_csv("model_1.csv")   # contains: time, E, T
df2 = pd.read_csv("model_2.csv")   # contains: time, E, T, I
df3 = pd.read_csv("model_3.csv")   # contains: time, E, T, N
```

```python
# Each contain E, and T, we will take final E, T capturing richer dynamics 
```

```python
# --------------------------------------------------------
#  Extract time, E, and T from each DataFrame
# --------------------------------------------------------
t1, E1, T1 = df1.iloc[:, 0], df1["E"], df1["T"]
t2, E2, T2 = df2.iloc[:, 0], df2["E"], df2["T"]
t3, E3, T3 = df3.iloc[:, 0], df3["E"], df3["T"]
```

```python
# --------------------------------------------------------
# Plot all E and T curves on the same axes
# --------------------------------------------------------
plt.figure(figsize=(12, 6), dpi = 200)

# Plot E values
ax1 = plt.subplot(1, 2, 1)
ax1.plot(t1, E1, label="E from Model 1")
ax1.plot(t2, E2, label="E from Model 2")
ax1.plot(t3, E3, label="E from Model 3")
ax1.set_title("Effector cells (E)")
ax1.set_xlabel("Time")
ax1.set_ylabel("E")
ax1.legend()
set_spines_black(ax1)
ax1.grid(True)

# Plot T values
ax2 = plt.subplot(1, 2, 2)
ax2.plot(t1, T1, label="T from Model 1")
ax2.plot(t2, T2, label="T from Model 2")
ax2.plot(t3, T3, label="T from Model 3")
ax2.set_title("Tumor cells (T)")
ax2.set_xlabel("Time")
ax2.set_ylabel("T")
ax2.legend()
set_spines_black(ax2)
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
# --- Extract time, E, T from Model 1 ---
time = df1.iloc[:, 0]     
E = df1["E"]
T = df1["T"]

# --- Extract I from Model 2 and N from Model 3 ---
I = df2["I"]
N = df3["N"]

# --- Combine into new DataFrame ---
x_train = pd.DataFrame({
    "time": time,
    "E": E,
    "T": T,
    "I": I,
    "N": N
})

# --- Save it for use ---
x_train.to_csv("combined_df.csv", index=False)
print(x_train.head())

```
